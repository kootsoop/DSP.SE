```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import csd,periodogram

t = np.arange(0, 2, 0.001);
x = np.sin(2*np.pi*200*t)
y = 3*np.sin(2*np.pi*200*(t-np.pi/2))
fig,ax = plt.subplots()
Cxy, f = ax.cohere(x, y)

plt.figure(2)
plt.plot(t, x)
plt.plot(t, y/3)
plt.xlim([0,0.1])

noise_1 = np.random.normal(0,0,len(x))
noise_2 = np.random.normal(0,0,len(x))
f,Pxx=csd(x , x )
f,Pyy=csd(y , y)
f,Pxy=csd(x , y)

Coh=np.zeros(len(f))
for i in range(len(f)):
    Coh[i]=abs(Pxy[i])**2/(abs(Pxx[i]*Pyy[i]))

fig,ax = plt.subplots()
ax.plot(f,Coh)
```
