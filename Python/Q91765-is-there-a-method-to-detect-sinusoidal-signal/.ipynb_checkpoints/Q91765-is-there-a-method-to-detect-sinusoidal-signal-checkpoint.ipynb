```python
import numpy as np
from matplotlib import pyplot as plt


N = 1024
M = 512

n0 = int(200*np.random.rand() + 100)

x = np.random.normal(size=N)
sinewave = np.cos(2*np.pi*np.linspace(0,10,M))
x[n0:(n0+M)] = x[n0:(n0+M)] + sinewave


periodogram = np.square(np.abs(np.fft.fft(x)))
omega_hat_index = np.argmax(periodogram)
omega_hat = 2*np.pi*omega_hat_index/N

alpha1_hat = 2/M*np.sum(x*np.cos(omega_hat*np.linspace(0,10*N/M,N)))
alpha2_hat = 2/M*np.sum(x*np.sin(omega_hat*np.linspace(0,10*N/M,N)))

Ahat = np.sqrt(np.square(alpha1_hat) + np.square(alpha2_hat))

plt.figure(1)
plt.plot(x)
plt.plot(np.linspace(n0,n0+M,M), sinewave)
plt.plot(np.cos(omega_hat*np.linspace(0,N,N)),'r')

plt.figure(2)
plt.plot(periodogram)
plt.plot(omega_hat_index,periodogram[omega_hat_index],'rx')






```
