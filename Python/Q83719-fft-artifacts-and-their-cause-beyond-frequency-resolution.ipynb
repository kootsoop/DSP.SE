```python
import numpy as np
import matplotlib.pyplot as plt

T = 100
np.random.seed(83719)
x = np.random.normal(0,1,T)
X = np.fft.fft(x)

Zero = np.argmax(np.abs(X[0:int(T/2)]))

window = np.ones(X.shape)
window[Zero] = 0
window[T-Zero] = 0

xx = np.real(np.fft.ifft(X * window))
WINDOW = np.real(np.fft.ifft(window))

plt.figure(1)
plt.plot(np.abs(X))
plt.plot(Zero,np.abs(X[Zero]),'ro')
plt.plot(T-Zero,np.abs(X[T-Zero]),'ro')
plt.title('Spectrum of $x$ and the points to be zeroed.')

plt.figure(2)
plt.plot(WINDOW)
plt.title('Effective filter response.')

plt.figure(3)
plt.plot(x)
plt.plot(xx)
plt.title('Original $x$ and the zeroed version.')

plt.figure(4)
plt.plot(x-xx)
plt.title('Difference between $x$ and the zeroed version.')

padding_factor = 10
X_zero_padded = np.fft.fft(x,T*padding_factor)
window_zero_padded = np.fft.fft(WINDOW, T*padding_factor)
xx_zero_padded = np.real(np.fft.ifft(X_zero_padded * window_zero_padded))

plt.figure(5)
plt.plot(np.abs(window_zero_padded))
plt.title('Window function zero padded.')

plt.figure(6)
plt.plot(x)
plt.plot(xx_zero_padded[0:T])
plt.title('Original $x$ and the zeroed version (zero padded).')

plt.figure(7)
plt.plot(x-xx_zero_padded[0:T])
plt.title('Difference between $x$ and the zeroed version (zero padded).')

```
