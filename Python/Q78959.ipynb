```python
import matplotlib.pyplot as plt
import numpy as np

T = 1000
fs = 44100
t = np.arange(T)/fs
omega = 2*np.pi*1000
phi = 2*np.pi*0.8978941234
x = 1 + np.sin(omega*t + phi)

plt.figure(1)
plt.plot(x)

Rxx = np.correlate(x,x,  mode='full')
plt.figure(2)
plt.plot(Rxx)

Rxx2 = np.correlate(x-np.mean(x), x-np.mean(x), mode='full')
plt.figure(3)
plt.plot(Rxx2)

Rxx2un = np.divide(Rxx2, np.bartlett(len(Rxx2)))
plt.figure(4)
plt.plot(Rxx2un)

y = np.pad(x,10)
Rxy = np.correlate(x-np.mean(x),y-np.mean(y), mode='full')
Rxyun = np.divide(Rxy, np.bartlett(len(Rxy)))
plt.figure(5)
plt.plot(Rxx2un)
plt.plot(Rxyun)

ax = plt.figure(6)
plt.plot(Rxx2un)
plt.plot(Rxyun)
plt.xlim(950,1050)
```
