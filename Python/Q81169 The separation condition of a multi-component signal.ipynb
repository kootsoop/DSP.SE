```python
import numpy as np
from scipy.signal import blackman
from matplotlib import pyplot as plt
import imageio

N = 64
omega1 = 2*np.pi*0.1
x1 = np.sin(omega1*np.arange(N))

images = []

figNo = 1
for omega2 in np.arange(N)/N*np.pi/4+2*np.pi*0.1:
    plt.figure(figNo)
    plt.plot(abs(np.fft.fft(x1)))
    x2 = np.sin(omega2*np.arange(N))
    plt.plot(abs(np.fft.fft(x2)))
    plt.plot(abs(np.fft.fft(x1+x2)))
    plt.ylim([0,50])
    plt.savefig('foo.png')
    images.append(imageio.imread('foo.png'))
    figNo = figNo + 1

imageio.mimsave('Q81169.gif', images)
```
