```python
import numpy as np
from matplotlib import pyplot as plt

def periodicity_check(signal):
    period_range = np.arange(1,len(signal))
    counts = np.zeros(len(period_range))
    for p in period_range:
        for jump in np.arange(1,p):
            counts[p-1] = counts[p-1] + np.mean(signal[jump:len(signal):p])
        counts[p-1] = counts[p-1]/p
        
    return period_range, counts

f = 0.03
x = np.sin(2*np.pi*np.arange(1,100)*f)
# x = np.random.normal(0,1,100)
p, cnt = periodicity_check(x)

idx = np.argmax(cnt)

plt.plot(x)
plt.plot(p, cnt/max(np.abs(cnt)))

print(str(p[idx]) + " vs " + str(1/f))
```
