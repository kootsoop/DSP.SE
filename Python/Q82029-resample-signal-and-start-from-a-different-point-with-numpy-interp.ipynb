```python
import matplotlib.pyplot as plt
import numpy as np

def resample_by_interpolation(sig, output_step, input_step, new_start_x = None ):

    scale = input_step / output_step
    # calculate new length of sample
    n = round(len(sig) * scale)
    
    if new_start_x is None:
        output_spacing = np.linspace(0, 1.0, n)
    else:
        if new_start_x > 1:
            print("new_start_x must be between 0 and 1")
            raise
        else:
            output_spacing = []
            current = new_start_x
            while (current < 1.0):
                output_spacing.append(current)
                current = current + output_step        

    resampled_sg = np.interp(
        output_spacing,  # where to interpret
        np.linspace(0.0, 1.0, len(sig)),  # known positions
        sig,  # known data points
    )
    return resampled_sg

T = 50
x = np.linspace(2, 8*np.pi, T)
y = np.sin(x)
output_step, input_step = ((x[1]-x[0])/2, x[1]-x[0])
new_y = resample_by_interpolation(y, output_step, input_step, new_start_x = 2.5 / T)
new_x = np.linspace(min(x), max(x), len(new_y))


plt.figure(1)
plt.plot(x, y, '--')
plt.plot(new_x, new_y, 'x')
plt.show()

plt.figure(2)
plt.plot(y)
plt.plot(new_y)
```
