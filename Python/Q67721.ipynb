```python
from numpy import linspace, pi, exp, mod
import matplotlib.pyplot as plt

def mod_2pi_central(value):
   return mod(value, 2*pi) - pi

plt.figure(figsize=(20,20))
mu = 0.2
phase = linspace(-pi, pi, 1000)
plt.plot(phase, exp(-mod_2pi_central(phase-mu)*mod_2pi_central(phase-mu)))
mu = pi
phase = linspace(-pi, pi, 1000)
plt.plot(phase, exp(-mod_2pi_central(phase-mu)*mod_2pi_central(phase-mu)))
plt.savefig('/Users/kootsoop/Pictures/Q67721.png')
```
