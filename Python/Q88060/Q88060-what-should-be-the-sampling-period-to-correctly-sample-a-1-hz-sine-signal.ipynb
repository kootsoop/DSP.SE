```python
import numpy as np
import matplotlib.pyplot as plt 

# Function definition
ti = 0.0 # [s]
tf = 8.0 # [s]
f0 = 1.0 # [Hz]
func = lambda t: np.cos(2*np.pi*f0*t)

# Sampling frequencies
# ws0 = 2*(2*np.pi*f0) 
ws0 = 1/0.001
ws1 = 2.1

# Sampling instants (vectors)
t0 = np.arange(ti, tf, 1/ws0)
t1 = np.arange(ti, tf, 1/ws1)

# Signal realizations
y0 = func(t0)
y1 = func(t1)

# Plots
fig, ax = plt.subplots()
ax.plot(t0, y0)
ax.plot(t1, y1, '.')
ax.grid()
```
