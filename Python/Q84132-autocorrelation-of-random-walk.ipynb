```python
import matplotlib.pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import acf

len_walk = 10000
n_walks  = 500

acf_sum = np.zeros(len_walk)

for j in range(n_walks):
  w = [0]
  for i in range(len_walk-1):
    w.append(w[-1] + np.random.normal())

  acf_result = acf(w, nlags=len_walk, fft=True)
  acf_sum = acf_sum + acf_result
  plt.plot(acf_result)  

plt.plot(acf_sum/n_walks, label='acf_sum', color="red",  linewidth=10)
```

```python
plt.plot(acf_sum/n_walks, label='acf_sum')
plt.grid()
plt.legend()
```

```python
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show

NN = 1000
s = np.arange(NN)+1
t =  np.arange(NN)+1

def cov_random_walk(t,s):
    return np.sqrt((t+1)/(s+1))


S,T = meshgrid(s, t) # grid of point
cov_rw = cov_random_walk(S, T) # evaluation of the function on the grid
im = imshow(cov_rw,cmap=cm.RdBu) # drawing the function
# adding the Contour lines with labels
cset = contour(cov_rw,np.arange(-10,10,0.5),linewidths=2,cmap=cm.Set2)
clabel(cset,inline=True,fmt='%1.1f',fontsize=10)
colorbar(im) # adding the colobar on the right
# latex fashion title
title('Some plot')
show()
```
