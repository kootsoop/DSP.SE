```python
import matplotlib.pyplot as plt
import numpy as np

T = 201
fs = 100
t = np.arange(T)/fs
x = np.sin(2*np.pi*2*t)
x_fft = np.fft.fft(x)
decimation_factor = 24

plt.figure(1)
plt.plot(t,x)

x_decimated = x[1:T:decimation_factor]
t_decimated = t[1:T:decimation_factor]

gain_factor = 6

plt.plot(t_decimated, x_decimated)

T_decimated = len(x_decimated)
T_decimated_on_2_plus_1 = int(T_decimated/2)+1
print(T_decimated)

x_d_fft = np.fft.fft(gain_factor*x_decimated)
x_d_fft_padded = np.concatenate((x_d_fft[0:T_decimated_on_2_plus_1], np.zeros(len(x)-len(x_d_fft)), x_d_fft[T_decimated_on_2_plus_1:T_decimated]))
x_interpolated = np.fft.ifft(gain_factor*x_d_fft_padded)
plt.figure(3)
plt.plot(t,x)
plt.plot(t, np.real(x_interpolated))
plt.plot(t, np.imag(x_interpolated))
plt.figure(4)
plt.plot(np.abs(np.fft.fft(x)))
```
