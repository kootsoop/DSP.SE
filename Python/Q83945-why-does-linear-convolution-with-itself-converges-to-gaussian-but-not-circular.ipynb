```python
import numpy as np
import matplotlib.pyplot as plt

def my_convolve(x,y,Nfft):
    X = np.fft.fft(x,Nfft)
    Y = np.fft.fft(y,Nfft)
    Z = X*Y
    return np.fft.ifft(Z,Nfft)

x = [1,1]

N = 20

xx = x.copy()

for idx in np.arange(N):
    xx = np.real(my_convolve(xx,x, 16))
    
plt.plot(xx)
```
